```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
from matplotlib import pyplot as plt
import os
import matplotlib.style
plt.style.use('classic')
%matplotlib inline
```

```python
df =pd.read_csv("/content/drive/MyDrive/SQL/sql_question1.csv")
```

```python
df
```

```python
# Plotting the bar graph
plt.figure(figsize=(10, 6))
bars = plt.bar(df['week_number'], df['crimes_reported'], color='skyblue', edgecolor='black')

# Adding counts on top of the bars
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # X-coordinate
        height,                            # Y-coordinate
        str(height),                       # The text (bar height)
        ha='center', va='bottom', fontsize=20  # Center align and font size
    )

# Adding titles and labels
plt.title('Crimes Reported by Week', fontsize=16)
plt.xlabel('Week Number', fontsize=12)
plt.ylabel('Crimes Reported', fontsize=12)
plt.xticks(df['week_number'], fontsize=10)
plt.yticks(fontsize=10)

# Display the graph
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

[Q6] What is the status of reported crimes?.

```python
df6=pd.read_csv("/content/drive/MyDrive/SQL/sql_question6.csv")
```

```python
df6
```

```python
# Plotting the pie chart
plt.figure(figsize=(6, 6))
plt.pie(
    df6['total_crimes'],
    labels=df6['case_status'],
    autopct='%1.1f%%',  # Display percentages
    startangle=90,      # Rotate the pie chart
    colors=plt.cm.Paired.colors,  # Optional: color palette
    wedgeprops={'edgecolor': 'black'}  # Add edge to wedges
)

# Adding a title
plt.title('Distribution of Crime Reports', fontsize=16)

# Display the pie chart
plt.tight_layout()
plt.show()
```

[Q9] Is crime more likely to be committed by relation of victims than strangers?

```python
df9=pd.read_csv("/content/drive/MyDrive/SQL/sql_question9.csv")
```

```python
df9
```

```python
# Aggregating the crime count by offender_relation
relation_totals = df9.groupby('offender_relation')['crime_count'].sum()

# Creating the pie chart
plt.figure(figsize=(6, 6))
plt.pie(
    relation_totals,
    labels=relation_totals.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['skyblue', 'orange'],
    wedgeprops={'edgecolor': 'black'}
)

# Adding title
plt.title('Total Crime Count by Offender Relation', fontsize=14)
plt.show()
```
